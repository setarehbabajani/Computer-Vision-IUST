```python
# Importing the mtm module and printing its version.
import mtm
print("mtm version: ", mtm.__version__)

# Importing specific functions from the mtm module.
from mtm import matchTemplates
from mtm.detection import plotDetections
```

# Showing the image

```python
# Importing necessary libraries
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os

# Define the path to the image
image_path = os.path.join('/content', 'coins.png')

# Load the image
image = mpimg.imread(image_path)

# Display the image in grayscale
plt.imshow(image, cmap='gray')
plt.axis('off')  # Hide the axis
plt.show()
```

# Generating template  



```python
small_Coin = image[37:37+38, 80:80+41]
large_Coin = image[14:14+59,302:302+65]
plt.figure(0)
plt.imshow(small_Coin, cmap="gray")
plt.figure(1)
plt.imshow(large_Coin, cmap="gray")
```

# Finding the coins of the image

```python
listTemplates = [small_Coin, large_Coin]
listDetections = matchTemplates(image, listTemplates, scoreThreshold=0.4, maxOverlap=0)
plotDetections(image, listDetections)
```
